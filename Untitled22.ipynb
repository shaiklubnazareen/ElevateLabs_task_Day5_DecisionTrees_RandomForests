```python

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
import graphviz
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
iris = load_iris()
X, y = iris.data, iris.target
feature_names = iris.feature_names
class_names = iris.target_names
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
dtree = DecisionTreeClassifier(max_depth=3, random_state=42)
dtree.fit(X_train, y_train)

```

```python
dot_data = export_graphviz(
    dtree,
    out_file=None,
    feature_names=feature_names,
    class_names=class_names,
    filled=True,
    rounded=True,
    special_characters=True
)
graph = graphviz.Source(dot_data)
graph.render("iris_decision_tree", format='png', cleanup=True)
graph.view()
```

```python
rforest = RandomForestClassifier(n_estimators=100, random_state=42)
rforest.fit(X_train, y_train)

```

```python
y_pred_tree = dtree.predict(X_test)
y_pred_rf = rforest.predict(X_test)

print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_tree))
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
importances = rforest.feature_importances_
forest_importances = pd.Series(importances, index=feature_names)

plt.figure(figsize=(8, 5))
forest_importances.sort_values().plot(kind='barh', color='skyblue')
plt.title("Feature Importance (Random Forest)")
plt.xlabel("Importance")
plt.tight_layout()
plt.show()
```

```python
cv_score_tree = cross_val_score(dtree, X, y, cv=5)
cv_score_rf = cross_val_score(rforest, X, y, cv=5)

print("\nCross-Validation Accuracy (Decision Tree):", np.round(cv_score_tree.mean(), 3))
print("Cross-Validation Accuracy (Random Forest):", np.round(cv_score_rf.mean(), 3))
```
